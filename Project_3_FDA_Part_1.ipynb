```python
#Strength Calculation - 
# Strength can be calculated by summing the weights rowwise or column wise
count = 0
df_result_set = pd.DataFrame(columns = ['Node Name','Strength','Degree'])
for i in df_adj_matrix.columns :
  degree = (df_adj_matrix[i] != 0).sum()
  df_result_set = df_result_set.append({'Node Name': i,'Strength': df_adj_matrix[i].sum(),'Degree': degree}, ignore_index=True)
```

```python
df_result_set
```

```python
# Top 10 Nodes by Strength
df_result_set[['Node Name','Strength']].sort_values(by=['Strength'],ascending=False).head(10)
```

```python
# Top 10 Nodes by Degree
df_result_set[['Node Name','Degree']].sort_values(by=['Degree'],ascending=False).head(10)
```

```python
df_weights = pd.DataFrame(columns = ['Node 1','Node 2','Weight'])
```

```python
for i in df_adj_matrix.index:
  for j in df_adj_matrix.columns:
    df_weights = df_weights.append({'Node 1': i,'Node 2': j,'Weight': df_adj_matrix[i][j]}, ignore_index=True)
```

```python
# Top 10 Node Pairs by Weight
df_weights.sort_values(by=['Weight'],ascending=False).head(10)
```
