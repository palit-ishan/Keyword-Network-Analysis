```python
# We will create a custom function to clean these tweets.
# Function for data cleaning
import re
import nltk 
def clean_tweets(tweet):
     
    # Cleaning tweets of punctuations
    tweet = re.sub('[()!?]', ' ', tweet)
    tweet = re.sub('\[.*?\]',' ', tweet)

     # Cleaning web links and hyperlinks
    tweet = re.sub(r'http\S+', ' ', tweet)
    tweet = re.sub(r"www.\S+", " ", tweet)
    
    # Cleaning hashtags and references
    tweet = re.sub("[@&#][A-Za-z0-9_]+"," ", tweet.lower())
   
    
    # Cleaning non alphanumeric characters in tweets
    tweet = re.sub("[^a-z0-9]"," ", tweet)
    
    # Cleaning numbers from tweets
    tweet = re.sub(r'[0-9]+', ' ', tweet)
    
    # Tokening at word level and removing stop words
    word_tokens = word_tokenize(tweet)
    filtered_sentence = [i for i in word_tokens if not i in stop_words]
    
    if len(filtered_sentence)==0:
        return float('NaN')
    else:
        
        return ' '.join(filtered_sentence)
    
```

```python
#We will create a custom function for calculating word frequencies using the nltk package
def word_frequency(document):
    k = nltk.FreqDist(document)
    return k
```

Zipf's Law - 
Written texts show the remarkable feature that the rank-ordered distribution of word frequencies follows an approximate power law - 
image.png

where r is the rank that is assigned to every word in the text. For most texts, regardless of language, time of creation, genre of literature, its purpose, etc. one finds that α ∼ 1, which is referred to as Zipf's law.
